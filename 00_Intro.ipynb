```python
from financetoolkit import Toolkit
import matplotlib.pyplot as plt
```

```python
api_key_file = './api_key.txt'
with open(api_key_file) as f:
    api_key = f.readline()
```

```python
companies = Toolkit(
    tickers=['005930.KS', '000660.KS'],
    api_key=api_key,
    start_date='2000-01-01'
)
```

```python
historical_data = companies.get_historical_data()
historical_data
```

```python
historical_data.keys()
```

```python
historical_data['Cumulative Return']
```

```python
historical_data['Cumulative Return'].plot(figsize=(15, 5), title='Cumulative Returns of Samsung & Hynix')
plt.show()
```

```python
close = historical_data['Adj Close']
close
```

```python
close.loc[:,'005930.KS'] = close['005930.KS'] / close['005930.KS'].iloc[0]
close.loc[:,'000660.KS'] = close['000660.KS'] / close['000660.KS'].iloc[0]
close.loc[:,'Benchmark'] = close['Benchmark'] / close['Benchmark'].iloc[0]
close
```

```python
close.plot(figsize=(15, 5), title='Normalized adj close of Samsung & Hynix')
plt.show()
```
